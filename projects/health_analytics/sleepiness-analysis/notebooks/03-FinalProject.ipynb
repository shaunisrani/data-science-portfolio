## Abstract

Our project’s goal was to establish whether or not there was a correlational relationship between sleep quality metrics and the development of neurodegenerative diseases such as Alzheimer's disease, Parkinson's disease, and dementia. In our research we chose to analyze Alzheimer's disease and Parkinson's disease. ## Research Question

To what extent can sleep quality metrics (sleep duration, sleep fragmentation, REM latency, and sleep efficiency) predict the risk of developing neurodegenerative diseases such as Alzheimer’s disease, Parkinson's disease, and dementia over time, accounting for demographic and lifestyle variables?


## Background and Prior Work

Sleep is one of the most important and necessary parts of our everyday routine, playing a critical role in helping us maintain cognitive function, brain health, memory, mood, emotional regulation, and overall well-being. It plays a vital role in the human health as a whole, allowing the brain to reset itself, remove accumulated toxic byproducts, and adjust neurotransmitter levels, enhancing learning by forming new neural connections and problem-solving abilities **[1]**. Research has shown there is a correlation between different processes that occur during our sleep and consequences with changes to sleep quality.

An emerging concern associated with sleep quality is the link between sleep and neurodegeneration, the loss of neurons connected with various behavioral impairments **[2]**. More generally, neurodegeneration can be seen as the progressive loss of neurons and functionalities. Sleep and sleep quality has become increasingly intertwined and linked to neurodegenerative diseases because of its large impact on brain functionality and regulation. For example, analyses have found that insomnia elevates the risk of all-cause dementia, especially when occurring with a major depressive disorder (MDD) **[3]**. This leads to our main research question: How much of an impact does sleep quality have on the likelihood of developing neurodegenerative disease or other similar health effects? According to studies, having a persistently short sleep duration at ages 50 to 70 compared to persistent normal sleep duration was associated with a 30% risk increase in dementia **[4]**. This was independent of sociodemographic, behavioural, cardiometabolic, and mental health factors as well. 

The amount of prior work done on this topic of sleep and neurodegeneration is pretty extensive as studies on our sleep and sleeping habits have a long and storied history. In the past few decades, researchers have made extensive progress on the study of sleep and its effects on all parts of the human body. Recently, researchers have recognized and identified sleep quality as a potential risk factor and symptom of neurodegeneration. However, many aspects of this relationship are still being studied presently and new data and relationships are being found.

### Parkinson's Progression Markers Initiative Dataset

### Import the data into 3 separate data frames
- ess_df will hold the data regarding the Epsworth Sleepiness Scale Questionaire
- rem_df will hold the data for the different types of sleep quality questions
- status_df will hold the patient numbers(PATNO) along with their cohort group and genetic mutations

```python
import pandas as pd
import os

# Load the data files (assuming they are in the same directory)
ess_df = pd.read_csv('data/00-raw/Epworth_Sleepiness_Scale_12Nov2025.csv')
rem_df = pd.read_csv('data/00-raw/REM_Sleep_Behavior_Disorder_Questionnaire_12Nov2025.csv')
status_df = pd.read_csv('data/00-raw/Participant_Status_12Nov2025.csv')
```

#### Prepare the ESS Data
The data inside the ESS sleep questionaire has multiple rows with the same PATNO because the questionaire is taken over time. To account for this and maintain one row for each patient in our data frame, we will only keep the first recorded date for each patient which will serve as a basline for all analysis. We label this initial row as **BL**.

The ESS file contains 8 columns which correspond to a question in the Epsworth sleepiness scale. The response to each question is a value of 0 - 3 which corresponds to the severity of drowsniness at a certain point in the day. 

```python
# 1. Filter for baseline visit
ess_baseline_df = ess_df[ess_df['EVENT_ID'] == 'BL'].copy()

# 2. Define the ESS columns for summation.
ess_cols = ['ESS1', 'ESS2', 'ESS3', 'ESS4', 'ESS5', 'ESS6', 'ESS7', 'ESS8']

# 3. Ensure columns are numeric (converts non-numeric values to NaN)
ess_baseline_df[ess_cols] = ess_baseline_df[ess_cols].apply(pd.to_numeric, errors='coerce')

# 4. Calculate the Total ESS score (summing all 8 items). 
ess_baseline_df['Total_ESS_Score'] = ess_baseline_df[ess_cols].sum(axis=1)

# 5. Calculate the average ESS score over time
ess_df[ess_cols] = ess_df[ess_cols].apply(pd.to_numeric, errors='coerce')
ess_df['Total_ESS_Score_visit'] = ess_df[ess_cols].sum(axis=1)
ess_avg = ess_df.groupby('PATNO', as_index=False)['Total_ESS_Score_visit'] \
                .mean() \
                .rename(columns={'Total_ESS_Score_visit': 'Avg_ESS_Score'})


# 5. Select only PATNO and the new Total_ESS score
ess_tidy = ess_baseline_df[['PATNO', 'Total_ESS_Score']].merge(ess_avg, on='PATNO', how='left')
```

#### Preparing the REM data

Similar to the ESS dataframe, we want to only keep the first recorded row in the data set for each patient. We label this baseline **BL** and filter the file to only keep these rows for the final data frame. 

Our target variables are PATNO, SLPDSTRB, MVAWAKEN, and DRMVIVID which correspond to the occurence of certain sleep behaviors amongst patients.

- SLPDSTRB = 0 or 1
- MVAAWAKEN = 0 or 1
- DRMVIVID = 0 or 1

```python
# 1. Filter for baseline visit
rem_baseline_df = rem_df[rem_df['EVENT_ID'] == 'BL'].copy()

# 2. Select the required REM sleep columns and PATNO
rem_cols = ['PATNO', 'SLPDSTRB', 'MVAWAKEN', 'DRMVIVID']
rem_tidy = rem_baseline_df[rem_cols]

# 3. Remove any residual duplicates, keeping the first entry (should be minimal after BL filter)
rem_tidy = rem_tidy.drop_duplicates(subset=['PATNO'], keep='first').astype(int)

# print("REM Tidy Data Head:")
rem_tidy.head()
```

#### Preparing the Genetics Data

Inside the participant status file, there are multiple columns corresponding to whether a patient has or does not have a certain genetic mutation. Each entry is between 0 and 1. 

We target only these genetic variables along with patient number and lastly account for any duplicates within the dataset to align with similar practices in previous steps.

- PATNO = int value starting from 3000
- COHORT = 1: Parkinsons, 2: Healthy Control, 3: SWEDD, 4: Prodromal
- ENROLL_AGE = float
- ENRLPINK1 - ENRLGBA = value of 0 or 1

```python
genetic_cols_to_keep = [
    'PATNO', 'ENROLL_AGE', 'COHORT', 'ENRLPINK1', 'ENRLPRKN', 'ENRLSRDC', 'ENRLNORM',
    'ENRLOTHGV', 'ENRLHPSM', 'ENRLRBD', 'ENRLLRRK2', 'ENRLSNCA', 'ENRLGBA'
]

status_tidy = status_df[genetic_cols_to_keep].copy()

# 1. Remove any duplicates just in case, keeping the first entry.
status_tidy = status_tidy.drop_duplicates(subset=['PATNO'], keep='first')

# print("Status/Genetics Tidy Data Head:")
status_tidy.head()
```

#### Merging Dataframes

We merge the 3 data frames with an outer join to preserve all the patients recorded in the participant status file. To further follow tidy data practices, we rename the 3 REM sleep metrics to more descriptive variable names. This will help in the readability and interpretibility of our analysis later on.

```python
# 1. Merge ESS and REM data
merged_df = pd.merge(ess_tidy, rem_tidy, on='PATNO', how='outer')

# 2. Merge with Status/Genetics data
final_tidy_df = pd.merge(merged_df, status_tidy, on='PATNO', how='outer')

# 3. Tidy Data Cleanup: Rename columns to be more descriptive, Reorder columns
final_tidy_df.rename(columns={
    'SLPDSTRB': 'REM_Sleep_Disturbance',
    'MVAWAKEN': 'Motor_Move_Waking',
    'DRMVIVID': 'Vivid_Dreams',
    'ENROLL_AGE': 'Age',
    'COHORT': 'Cohort',
    'PATNO': 'Patient_Number'
}, inplace=True)

cols = final_tidy_df.columns.tolist()
new_order = ['Patient_Number', 'Age', 'Cohort'] + [c for c in cols if c not in ('Patient_Number', 'Age', 'Cohort')]
final_tidy_df = final_tidy_df[new_order]
cols

# Save the final DataFrame to a CSV file
output_dir = 'data/02-processed'
os.makedirs(output_dir, exist_ok=True)
final_out_file = os.path.join(output_dir, 'parkinsons_tidy_data.csv')
final_tidy_df.to_csv(final_out_file, index=False)

print("Final Tidy DataFrame Head:")
final_tidy_df.head()
```

### NACC Alzheimer's x Sleep Factors Dataset

#### Data Cleaning and Merging Steps

- Summary of Data Cleaning & Processing Steps
1. Variable Selection: We started with the massive raw NACC dataset (potentially 2,000+ columns) and filtered it down to a targeted subset of ~56 variables. These variables were chosen based on our hypothesis to capture:
    - Predictors: Sleep metrics (NPI-Q, Mayo Sleep Questionnaire, SCOPA-Sleep).
    - Outcomes: Clinical diagnoses (AD, LBD, PD), cognitive scores (MMSE, MoCA), and biomarkers (CSF, Autopsy).
    - Covariates: Demographics, genetics (APOE), and medical history.
2. "Junk Code" Cleaning: The NACC dataset uses specific integers to represent missing or inapplicable data. We replaced the following codes with NaN (Not a Number) to prevent statistical errors, these are:
    - -4: Not available / form not submitted.
    - 9, 99, 999: Unknown.
    - 8, 88: Not applicable / Not assessed.
    - Note: Applied this replacement only to numeric columns as needed to avoid breaking text identifiers like NACCID.
3. Cohort Filtering (Row Selection): To test our specific hypothesis, we filtered the participants to include only three distinct groups:
    - Alzheimer's: Primary Etiology (NACCETPR) = 1.
    - Lewy Body / Parkinson's: Primary Etiology (NACCETPR) = 2.
    - Normal Controls: Cognitive Status (COGSTAT) = 2.
    - Note: This step automatically excluded participants with Frontotemporal Dementia (FTD) or other primary causes to reduce noise and cloud analyses with differing biomarkers

- Through feature engineering, we were able to create a single, clean Diagnosis column to label these three groups (Alzheimers, Lewy Body / PD, Normal Control). This makes plotting and statistical comparison with techniques like ANOVA, clustering, and regression much easier.
- Through type conversion, we converted categorical variables (like SEX, RACE, NACCAPOE) from integers into categorical types (factors) so Python treats them as groups rather than numbers, without breaking text identifiers.

```python
import pandas as pd
import numpy as np

# 1. Load the dataset
my_data = pd.read_csv('data/00-raw/NACC_data.csv')

# 2. Define the "Shopping List" of variables to keep
variables_to_keep = [
    # Identifiers
    "NACCID", "VISITYR", "NACCETPR", 
    
    # Demographics & Covariates
    "NACCAGE", "SEX", "EDUC", "RACE", "NACCINCM",
    "NACCAPOE", "NACCNE4S", "FADMOM", "FADPOP",
    "TOBAC30", "SMOKYRS", "ALCOHOL", "NACCACT",
    "CVHATT", "CVAFIB", "CVCHF", "CVSTROKE", "DIABETES", "DEPRS",
    
    # Sleep Metrics (Predictors)
    "NPIQINF", "NITE", "HYPOSOM", "RBD", 
    "LBSPDRM", "LBSPTALK", "LBSPMOVE", "LBSCDRM",
    "CONSFALL", "CONSWKOF", "CONSLYAW", "CONSWKER", "CONSLTTL", "CODSAWDY",
    
    # Outcomes (Diagnoses & Scores)
    "COGSTAT", "NACCALZD", "PARK", "NACCLBDS", "FTD",
    "NACCMMSE", "NACCMOCA",
    
    # Motor Exams (UPDRS)
    "LBUMSPCH", "LBUMFACE", "LBUMREST", "LBUMACT", "LBUMRIG", 
    "LBUMFALL", "LBUMGAIT", "LBUMPOST", "LBUMBRAD",
    
    # Biomarkers
    "NACCACSF", "NACCPCSF", "NACCTCSF", "NACCLEWY", "NPLBOD"
]

# 3. Create the analysis dataframe
# We intersect the list with columns that actually exist to avoid errors
existing_vars = [col for col in variables_to_keep if col in my_data.columns]
analysis_data = my_data[existing_vars].copy()

# 4. Clean "Junk" Codes
# Replace -4, 8, 9, etc. with NaN (only in numeric columns)
junk_codes = [-4, 8, 9, 88, 99, 999]

# Iterate over numeric columns and replace junk codes
numeric_cols = analysis_data.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    analysis_data[col] = analysis_data[col].replace(junk_codes, np.nan)

# 5. Filter Rows: Keep AD, LBD/PD, and Normal Controls
# Logic: Keep if Primary Diagnosis is AD (1), LBD (2), OR if Cognitive Status is Normal (2)
analysis_data = analysis_data[
    (analysis_data['NACCETPR'] == 1) | 
    (analysis_data['NACCETPR'] == 2) | 
    (analysis_data['COGSTAT'] == 2)
].copy()

# 6. Create the clean 'Diagnosis' column
conditions = [
    analysis_data['NACCETPR'] == 1,
    analysis_data['NACCETPR'] == 2,
    analysis_data['COGSTAT'] == 2
]
choices = ['Alzheimers', 'Lewy Body / PD', 'Normal Control']

analysis_data['Diagnosis'] = np.select(conditions, choices, default='Unknown')

# 7. Convert categorical columns to 'category' type (Factor equivalent)
categorical_cols = [
    "SEX", "RACE", "NACCAPOE", "NACCNE4S", "TOBAC30", "ALCOHOL",
    "CVHATT", "CVAFIB", "CVCHF", "DIABETES", "NPIQINF", "NITE", "HYPOSOM", "RBD",
    "LBSPDRM", "LBSPMOVE", "LBSCDRM", "CONSFALL", "CONSWKOF", "CONSLYAW", 
    "CONSWKER", "CONSLTTL", "CODSAWDY", "COGSTAT", "NACCALZD", "PARK", 
    "NACCLBDS", "FTD", "NACCLEWY", "NPLBOD", "Diagnosis"
]

# Only convert columns that actually exist in the filtered data
existing_cat_cols = [col for col in categorical_cols if col in analysis_data.columns]
analysis_data[existing_cat_cols] = analysis_data[existing_cat_cols].astype('category')

# The final check: 
print("Final Data Dimensions:", analysis_data.shape) # dimensions of dataframe
print("\nCounts per Diagnosis Group:")
print(analysis_data['Diagnosis'].value_counts())
print("\nFirst 5 rows:")
display(analysis_data.head())
```

### Analyzing REM Sleep metrics in the Parkinson's Dataset

We are analyzing 3 different REM sleep metrics: general sleep disurbance, vivid dreams, and motor move waking. These three variables contribute to the overall sleep quality of each patient and could serve as a good indicator of whether we can reject our null hypothesis. Specifically, general sleep disturbances is a variable for overall sleep disruptions that inhibit normal sleep, vivid dreams is a indicator of REM behavior disorder, and motor move waking could be a byproduct of nightmares or involuntary movements that lead to sleep disruption. All together, these metrics help to illustrate whether a patient has healthy sleep quality.

For each variable, let us plot the percentage of people within in each cohort that say they experience each.


```python
def plot_rem_var(var):
    df = park_df.copy()

    df[var] = pd.to_numeric(df[var], errors="coerce")

    # Compute percentages per cohort
    pct = df.groupby("Cohort")[var].mean() * 100
    pct_df = pct.reset_index()
    pct_df.columns = ["Cohort", "Percent"]

    plt.figure(figsize=(6, 5))

    # Match style of your countplot call
    ax = sns.barplot(
        data=pct_df,
        x="Cohort",
        y="Percent",
        hue="Cohort",
        palette="viridis"
    )

    labels = ['Parkinsons', 'Healthy Control', 'SWEDD', 'Prodromal']

    handles, _ = ax.get_legend_handles_labels()
    ax.legend(handles, labels, title="Cohort")
    plt.title(var.replace("_", " "))
    plt.xlabel("Cohorts")
    plt.ylabel("Percent with Behavior")
    plt.ylim(0, 100)

    # Annotate bars
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(
            f"{height:.1f}%",
            (p.get_x() + p.get_width() / 2, height),
            ha="center", va="bottom"
        )

    plt.show()
```


The healthy control group has the lowest proportion of patients experiencing REM sleep disturbances. There is an outlier for SWEDD which holds the highest proportion. However, earlier we discussed how this group accounts for less than 1 percent of the overall population in our dataset so it's results could have little impact on our overall conclusions.

Next, lets look at whether gentic mutations and sleep are correlated?

```python
# REM vars and candidate genetic columns
rem_vars = ["REM_Sleep_Disturbance", "Motor_Move_Waking", "Vivid_Dreams"]
genetic_cols = [
    "ENRLPINK1", "ENRLPRKN", "ENRLSRDC", "ENRLNORM",
    "ENRLOTHGV", "ENRLHPSM", "ENRLRBD", "ENRLLRRK2",
    "ENRLSNCA", "ENRLGBA"
]
subset_cols = rem_vars + genetic_cols

rem_df = park_df.dropna(subset=subset_cols)

# keep only columns that exist in the dataframe
genetic_cols = [c for c in genetic_cols if c in rem_df.columns]
rem_vars = [r for r in rem_vars if r in rem_df.columns]

# create any_mutation flag: 1 if any listed gene == 1, else 0
gdf = rem_df[genetic_cols].apply(pd.to_numeric, errors="coerce").fillna(0).astype(int)
rem_df["any_mutation"] = (gdf.sum(axis=1) > 0).astype(int)

# build numeric dataframe for correlation (fill NA with 0 for binary flags)
corr_cols = rem_vars + genetic_cols + ["any_mutation"]
corr_df = rem_df[corr_cols].apply(pd.to_numeric, errors="coerce").fillna(0)

# correlation matrix 
corr = corr_df.corr()

# REM vars vs any_mutation + individual genes
plot_cols = genetic_cols + ["any_mutation"]
subcorr = corr.loc[rem_vars, plot_cols]

plt.figure(figsize=(len(plot_cols)*0.6 + 2, len(rem_vars)*1.2 + 1))
sns.heatmap(subcorr, annot=True, fmt=".2f", cmap="vlag", center=0)
plt.title("Correlation (Pearson) between REM flags and gene mutations")
plt.xlabel("Gene / Any mutation")
plt.ylabel("REM variable")
plt.tight_layout()
plt.show()
```

In the correlation matrix, we examine that there is insiginficant correlation between gene mutations and the three REM features we are analyzing except for patients who have REM sleep behavior disorder. This indicates that individuals who have the preexisting condition of RBD are more likely to experience REM sleep disturbance, motor movements during sleep that leads to waking, and vivid dreams.

```python
genetic_cols = [c for c in [
    "ENRLPINK1","ENRLPRKN","ENRLSRDC","ENRLNORM",
    "ENRLOTHGV","ENRLHPSM","ENRLRBD","ENRLLRRK2",
    "ENRLSNCA","ENRLGBA"
] if c in park_df.columns]

gene_df = park_df.copy()

if "any_mutation" not in gene_df.columns:
    gdf = gene_df[genetic_cols].apply(pd.to_numeric, errors="coerce").fillna(0).astype(int)
    gene_df["any_mutation"] = (gdf.sum(axis=1) > 0).astype(int)


# make Cohort numeric and map labels
gene_df["Cohort"] = pd.to_numeric(gene_df["Cohort"], errors="coerce").astype("Int64")
label_map = {1: "Parkinsons", 2: "Healthy Control", 3: "SWEDD", 4: "Prodromal"}

# summary table: counts and percent with any mutation per cohort
summary = gene_df.groupby("Cohort").agg(
    n_total=("any_mutation", "count"),
    n_with=("any_mutation", "sum")
).reset_index()
summary["pct_with"] = summary["n_with"] / summary["n_total"] * 100
summary["Cohort_label"] = summary["Cohort"].map(label_map).fillna(summary["Cohort"].astype(str))

print(summary[["Cohort_label", "n_with", "n_total", "pct_with"]])

# bar plot: percent with mutation by cohort
plt.figure(figsize=(8,4))
sns.barplot(data=summary, x="Cohort_label", y="pct_with", palette="viridis")
plt.ylim(0, 100)
plt.ylabel("Percent with any genetic mutation")
plt.title("Prevalence of any genetic mutation by Cohort")
for i, row in summary.iterrows():
    plt.text(i, row.pct_with + 1, f"{row.pct_with:.1f}%", ha="center")
plt.tight_layout()
plt.show()
```

Judging from the graph, another confounding variable we may not have accounted for are genetic mutations within the patients. In this graph it shows the prevalence of any patient that has at least one genetic mutation. This information could lead us to believe that parkinsons could be the result of underlying genetic mutations instead of sleep. However, we can not fully discount sleep quality as an indicator for parkinsons because it could be that sleep further contributes to the development of parkinsons alongside genetic mutations.

This systematic gap limits the ability to link sleep metrics directly to physiological AD pathology and forces a stronger reliance on clinical diagnoses instead. 

Now, let's proceed with logistic regression to quantify risk in clinical data. We will use the statsmodel library since it provides detailed statistical outputs (p-values, confidence intervals) and allows us to easily calcualte Odds Ratios (OR), the standard metric for reporting risk in medical studies.

For our modeling strategy, we will filter the data to compare the Alzheimer's (Case) vs. Normal Control (Reference). Excluding the LBD/PD group for this specific model to get a clear signal on AD risk, we set our target variable is_AD (1 = Alzheimer's, 0 = Normal Control). Our predictors are the following: 1. Sleep Metrics: NITE (Nighttime behaviors), HYPOSOM (Insomnia), CODSAWDY (Daytime Sleepiness). 2. Controls/Covariates: NACCAGE (Age), SEX (Sex), DEPRS (Depression - Crucial confounder we indentified in EDA) -> dont have data for, removed. 

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# 1. Prepare the Data
# Filter for only AD and Normal Controls
model_df = nacc_df[nacc_df['Diagnosis'].isin(['Alzheimers', 'Normal Control'])].copy()

# Create Binary Target: 1 for Alzheimer's, 0 for Normal Control
model_df['is_AD'] = (model_df['Diagnosis'] == 'Alzheimers').astype(int)

# Ensure numeric types for regression
# REMOVED 'DEPRS' from this list
cols_to_convert = ['is_AD', 'NACCAGE', 'NITE', 'HYPOSOM', 'CODSAWDY']
model_df[cols_to_convert] = model_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values in these specific columns
model_df = model_df.dropna(subset=cols_to_convert)

print(f"N for Model: {len(model_df)} observations")

# 2. Define and Fit the Logistic Regression Model
# REMOVED 'DEPRS' from the formula
model_formula = 'is_AD ~ NITE + HYPOSOM + CODSAWDY + NACCAGE + C(SEX)'

logit_model = smf.logit(formula=model_formula, data=model_df).fit()

# 3. Print Summary
print(logit_model.summary())

# 4. Calculate and Interpret Odds Ratios (OR)
odds_ratios = pd.DataFrame({
    'Odds Ratio (Risk Factor)': np.exp(logit_model.params),
    'P-Value': logit_model.pvalues,
    '2.5% CI': np.exp(logit_model.conf_int()[0]),
    '97.5% CI': np.exp(logit_model.conf_int()[1])
})

print("\n--- RISK QUANTIFICATION (Odds Ratios) ---")
display(odds_ratios)
```

Here, we have our results showing the relationship between sleep and Alzheimer's without controlling for depression. Looking at the Odds Ratios table and the model summary generated: 

1. The most important metric is the Odds Ratio (Risk Factor) for our sleep variables (NITE, HYPOSOM, CODSAWDY).
